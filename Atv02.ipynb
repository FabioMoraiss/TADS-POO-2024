</p>

- Observação: o valor do desconto é calculado por meio da fórmula: 
    valor do desconto = valor do produto * percentual de desconto / 100.

```java
import java.util.Scanner;

Scanner input = new Scanner(System.in);

        System.out.println("digite o valor do produto");
        double valorProdutoSemDesconto = input.nextDouble();

        System.out.println("digite o valor do desconto em %");
        double descontoEmPorcentagem = input.nextDouble();

        double descontoNumerico =  (1-(descontoEmPorcentagem / 100));
        double valorProdutoComDesconto = valorProdutoSemDesconto - descontoNumerico;

        System.out.println("valor com desconto: " +valorProdutoComDesconto);
        System.out.println("valor DO desconto: " + descontoNumerico);

```
